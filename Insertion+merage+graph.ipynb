```python
# insertion algo
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr
```

```python
#test
print(insertion_sort([5, 3, 1, 4]))
```

```python
#avg time
def average_time(arr, repetitions=10):
    total_time = 0

    for _ in range(repetitions):
        temp = arr.copy()

        start = time.perf_counter()
        insertion_sort(temp)
        end = time.perf_counter()

        total_time += (end - start)

    average = total_time / repetitions
    return average * 1000
```

```python

df = pd.read_csv("clean_passenger_id.csv")

data = df["Passenger ID"].tolist()

print(data[:10])



# Merge Sort Algorithm

def merge_sort(arr):

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):

    result = []

    i = 0
    j = 0

    while i < len(left) and j < len(right):

        if left[i] < right[j]:

            result.append(left[i])
            i += 1

        else:

            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result

#test

print(merge_sort([5, 3, 1, 4]))





def average_time(arr, repetitions=10):

    total_time = 0

    for _ in range(repetitions):

        temp = arr.copy()

        start = time.perf_counter()

        merge_sort(temp)

        end = time.perf_counter()

        total_time += (end - start)

    average = total_time / repetitions

    return average * 1000



# input size

data_500 = data[:500]
data_1000 = data[:1000]
data_1500 = data[:1500]
data_2000 = data[:2000]
data_2500 = data[:2500]
data_3000 = data[:3000]
data_3500 = data[:3500]
data_4000 = data[:4000]
data_4500 = data[:4500]
data_5000 = data[:5000]
data_5500 = data[:5500]
data_6000 = data[:6000]
data_6500 = data[:6500]
data_7000 = data[:7000]
data_7500 = data[:7500]
data_8000 = data[:8000]
data_8500 = data[:8500]
data_9000 = data[:9000]
data_9500 = data[:9500]
data_10000 = data[:10000]
data_10500 = data[:10500]
data_11000 = data[:11000]





print("500:", len(data_500))
print("1000:", len(data_1000))
print("1500:", len(data_1500))
print("2000:", len(data_2000))
print("2500:", len(data_2500))
print("3000:", len(data_3000))
print("3500:", len(data_3500))
print("4000:", len(data_4000))
print("4500:", len(data_4500))
print("5000:", len(data_5000))
print("5500:", len(data_5500))
print("6000:", len(data_6000))
print("6500:", len(data_6500))
print("7000:", len(data_7000))
print("7500:", len(data_7500))
print("8000:", len(data_8000))
print("8500:", len(data_8500))
print("9000:", len(data_9000))
print("9500:", len(data_9500))
print("10000:", len(data_10000))
print("10500:", len(data_10500))
print("11000:", len(data_11000))



# average

print("\nAverage Execution Time (ms)\n")

print("500:", average_time(data_500))
print("1000:", average_time(data_1000))
print("1500:", average_time(data_1500))
print("2000:", average_time(data_2000))
print("2500:", average_time(data_2500))
print("3000:", average_time(data_3000))
print("3500:", average_time(data_3500))
print("4000:", average_time(data_4000))
print("4500:", average_time(data_4500))
print("5000:", average_time(data_5000))
print("5500:", average_time(data_5500))
print("6000:", average_time(data_6000))
print("6500:", average_time(data_6500))
print("7000:", average_time(data_7000))
print("7500:", average_time(data_7500))
print("8000:", average_time(data_8000))
print("8500:", average_time(data_8500))
print("9000:", average_time(data_9000))
print("9500:", average_time(data_9500))
print("10000:", average_time(data_10000))
print("10500:", average_time(data_10500))
print("11000:", average_time(data_11000))
```
